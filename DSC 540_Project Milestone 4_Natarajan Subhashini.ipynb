```python
# Import required libraries
import os
import requests
import json 
import pandas as pd
```

```python
# Call the API
url='https://open.er-api.com/v6/latest/USD'
resp =requests.get('https://open.er-api.com/v6/latest/USD')
```

```python
#Check the API response
print(resp.status_code)
```

```python
#Store the API response in a variable
data=resp.text
```

```python
# Verify data
print(data)
```

```python
# Parse the json data
parse_json=json.loads(data)
parse_json
```

```python
#  Store the rates variable from json
forexrates_json = parse_json['rates']
forexrates_json
```

```python
# Load the json dictionary into a dataframe
forex_latest=pd.DataFrame(forexrates_json.items(), columns=['Currency', 'ExchangeRate'])
forex_latest
```

```python
# Add a field in the dataframe to capture load date
from datetime import date
forex_latest['Date']=date.today()
forex_latest
```

```python
# Read an excel file with Currency code and country name
countryCurrency=pd.read_excel("list_of_currency_in_world_by_country-750j.xlsx")
countryCurrency
```

```python
# Join the two dataframes
forex = pd.merge(forex_latest, countryCurrency, how='inner', left_on = 'Currency', right_on = 'CurrencyCodeA')
forex= forex[['Currency', 'ExchangeRate','CountryName','CurrencyName']]
forex
```

```python
# Transform the country name and currency name data into upper case
forex.CountryName = forex['CountryName'].str.upper()
forex.CurrencyName = forex['CurrencyName'].str.upper()
forex
```

```python
# Remove duplicates in the dataframe
forex = forex.drop_duplicates()
forex
```

```python
# Identify number of countries using the same currency
currency_distro = forex.groupby(["Currency"]).count().reset_index()
```

```python
# Rename the column names
currency_distro= currency_distro[['Currency', 'ExchangeRate']]
currency_distro.columns =['Currency', 'NumberOfCountries']
currency_distro
```

```python
# Currencies used in more than one country
currency_distro = currency_distro[currency_distro["NumberOfCountries"]>1]
currency_distro
```

```python
# Plot the forex rates using pyplot

import matplotlib.pyplot as plt

plt.barh(currency_distro.Currency, currency_distro.NumberOfCountries, color='green')
plt.xlabel("Currency")
plt.ylabel("Number of Countries with the currency")
plt.title("Currency used in more than one country")

plt.show()
```
